```python
import torch

inputs = torch.tensor(
  [[0.43, 0.15, 0.89], # Your     (x^1)
   [0.55, 0.87, 0.66], # journey  (x^2)
   [0.57, 0.85, 0.64], # starts   (x^3)
   [0.22, 0.58, 0.33], # with     (x^4)
   [0.77, 0.25, 0.10], # one      (x^5)
   [0.05, 0.80, 0.55]] # step     (x^6)
)
```

First finding dot product of each vector with other vectors and it is known as attention score

```python
atten_scores= inputs @ inputs.T
print(atten_scores)
```

Now finding attention weight using pytorch soft-max which is also care of overflow and underflow of extreme values
</br>
meaning of dim = -1 is calcute soft max column wise
<br>
If attn_scores is a 2D tensor (for example, with a shape of [rows, columns]), dim=-1 will normalize across the columns so that the values in each row (summing over the column dimension) sum up to 1.

```python
atten_weight= torch.softmax(atten_scores,dim=-1)
print(atten_weight)
```

```python
row_2_sum = sum([0.1385, 0.2379, 0.2333, 0.1240, 0.1082, 0.1581])
print("Row 2 sum:", row_2_sum)
print("All row sums:", atten_weight.sum(dim=-1))
```

Now calculate attention vector

```python
all_context_vecs = atten_weight @ inputs
print(all_context_vecs)
```
